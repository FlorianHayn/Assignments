```python
#FIB
months = int(input())
offspring = int(input())

small = 1
medium = 0
rep = 0

for i in range(1, months):
    rep += medium
    offspring = rep * 3
    medium = small
    small = offspring

print(small + medium + rep)
```

```python
#FIBD
duration = int(input())
age_lim = int(input())

rabbits = [1]

for i in range(1, age_lim):
    rabbits.append(0)
print(rabbits)
for n in range(duration - 1):
    reprodRabbits = 0
    for j in range(1, len(rabbits)):
        reprodRabbits += rabbits[j]
    for m in range(1, len(rabbits)):
        rabbits[-m] = rabbits[-(m + 1)]
    rabbits[0] = reprodRabbits
    print(rabbits)
print(sum(rabbits))
```

```python
#LCSM
#About 2.5 minutes runtime :s
from Bio import SeqIO

sequences = []
ids = []

for x in SeqIO.parse("lcsm2.txt", "fasta"):
    sequences.append(x.seq)
    ids.append(x.id)

all_substrings = [sequences[0][i: j] for i in range(len(sequences[0]))
          for j in range(i + 1, len(sequences[0]) + 1)]

all_substrings.sort()
all_substrings.sort(key=len, reverse=True)

for i in range(len(all_substrings)):
    checked_sub = all_substrings[i]
    score = 0
    for n in range(1, len(sequences)):
        if checked_sub in sequences[n]:
            score += 1
    if score == len(sequences) - 1:
        print(all_substrings[i])
        break
```

```python
#REVP
def check_rev_comp(seq):
    revComp = []
    for element in seq:
        if element == "A":
            revComp.insert(0, "T")
        elif element == "T":
            revComp.insert(0, "A")
        elif element == "G":
            revComp.insert(0, "C")
        elif element == "C":
            revComp.insert(0, "G")
    result = "".join(revComp)
    if seq == result:
        return(True)

from Bio import SeqIO
sequences = []

for x in SeqIO.parse("revp.txt", "fasta"):
    sequences.append(x.seq)

dna_seq = sequences[0]

for i in range(len(dna_seq)):
    for n in range(4, 13):
        if n + i > len(dna_seq):
            break
        checked_seq = dna_seq[i:i+n]
        if check_rev_comp(checked_seq) == True:
            print(i + 1, n)
```

```python
#LGIS - This solution produces strictly de/-increasing sequences of the correct length. Rosalind however does not accept
#my solutions. I don't know what is wrong at this point.
def lengthLIS(numbers):
    LIS = [1] * len(numbers)

    for i in range(len(numbers) - 1, -1, -1):
        for j in range(i + 1, len(numbers)):
            if numbers[i] < numbers[j]:
                LIS[i] = max(LIS[i], 1 + LIS[j])
    return max(LIS), LIS

def seqLIS(numbers, LIS, lengthLIS):
    counter = lengthLIS
    seq_LIS = []
    while counter > 0:
        for i in range(len(LIS)):
            if counter == LIS[i]:
                seq_LIS.append(numbers[i])
                counter -= 1
                break
    return seq_LIS

def lengthLDS(numbers):
    LDS = [1] * len(numbers)

    for i in range(len(numbers) - 1, -1, -1):
        for j in range(i + 1, len(numbers)):
            if numbers[i] > numbers[j]:
                LDS[i] = max(LDS[i], 1 + LDS[j])
    return max(LDS), LDS

def seqLDS(numbers, LDS, lengthLDS):
    counter = lengthLDS
    seq_LDS = []
    while counter > 0:
        for i in range(len(LDS)):
            if counter == LDS[i]:
                seq_LDS.append(numbers[i])
                counter -= 1
                break
    return seq_LDS

txt_file = open("lgis.txt", "r")

n = int(txt_file.readline().strip())
numbers = list(map(int, txt_file.readline().split()))

txt_file.close()

abs_lengthLIS, globalLIS = lengthLIS(numbers)
abs_lengthLDS, globalLDS = lengthLDS(numbers)

print(*seqLIS(numbers, globalLIS, abs_lengthLIS))
print(*seqLDS(numbers, globalLDS, abs_lengthLDS))
```
