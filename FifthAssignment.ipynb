```python
#ORF
from Bio import SeqIO
from Bio.Seq import Seq

dictionary = {}

def fastaReader(fasta_file):
    for x in SeqIO.parse(fasta_file, "fasta"):
        dictionary[x.name] = (x.seq)
        return dictionary

ORFset = set()

def ORFGen(sequence):
    for i in range(len(sequence)):
        currentORF = []
        if sequence[i] == "M":
            for n in range(i, len(sequence)):
                currentORF.append(sequence[n])
                if sequence[n] == "*":
                    printseq = "".join(currentORF)
                    printseq = printseq[:-1]
                    ORFset.add(printseq)
                    break
                else:
                    continue

fastaReader("orftest.txt")

for element in dictionary:
    sequence = dictionary[element]
    rev_com_seq = sequence.reverse_complement()

raw_trans_forward_0 = sequence.translate()
ORFGen(raw_trans_forward_0)

raw_trans_forward_1 = sequence[1:-2].translate()
ORFGen(raw_trans_forward_1)

raw_trans_forward_2 = sequence[2:-1].translate()
ORFGen(raw_trans_forward_2)

raw_trans_reverse_0 = rev_com_seq.translate()
ORFGen(raw_trans_reverse_0)

raw_trans_reverse_1 = rev_com_seq[1:-2].translate()
ORFGen(raw_trans_reverse_1)

raw_trans_reverse_2 = rev_com_seq[2:-1].translate()
ORFGen(raw_trans_reverse_2)

ORFset = list(ORFset)
for i in range(len(ORFset)):
    print(ORFset[i])
```

```python
#PPER
from math import factorial

def binomial(n,m):
	return factorial(n) / (factorial(m) * factorial(n - m))

f = open("rosalind_pper.txt")

[n, m] = map(int, f.readline().split('\n')[0].split(' '))

f.close()

print(binomial(n, m) * factorial(m) % 1000000)
```

```python
#LONG
from Bio import SeqIO

def fastaReader(fasta_file):
    seq = []
    for x in SeqIO.parse(fasta_file, "fasta"):
        seq.append(str(x.seq))
    return seq

def overlap_func(s1, s2):

    fused = []
    overlapped = []
    
    for i in range(len(s1)):
        if s1[i:] == s2[:len(s1)-i]:
            overlapped.append(s1[i:])
            fused.append(s1+s2[len(s1)-i:])
            break
    for i in range(len(s2)):
        if s2[i:] == s1[:len(s2)-i]:
            overlapped.append(s2[i:])
            fused.append(s2+s1[len(s2)-i:])
            break
    if not overlapped:
        return "", ""

    fused_string = min(fused, key=len)
    
    overlapped_string = max(overlapped, key=len)
    
    return fused_string, overlapped_string

def find_superstring(strings):
    result = strings

    while len(result) > 1:
        most_overlapping_string = ""
        most_overlapping_string_pair = []
        most_overlapping_string_length = 0

        for i in range(len(result)-1):
            for j in range(i+1, len(result)):
                combine_string, overlap_string = overlap_func(result[i], result[j])
                if len(overlap_string) > most_overlapping_string_length:
                    most_overlapping_string = combine_string
                    most_overlapping_string_pair = [result[i], result[j]]
                    most_overlapping_string_length = len(overlap_string)

        result.remove(most_overlapping_string_pair[0])
        result.remove(most_overlapping_string_pair[1])
        result.append(most_overlapping_string)
        
    return result 

seq_string = []
seq_string = fastaReader("rosalind_long.txt")
superstring = find_superstring(seq_string)
print("".join(superstring))
```

```python
#TREE
data = "rosalind_tree.txt"

with open(data, "r") as f:
    n = int(f.readline())
    adjacency_list = [line.strip().split(" ") for line in f]

subtrees = []
nodes = set()
for i, j in adjacency_list:
    if i in nodes or j in nodes:
        for st in subtrees:
            if i in st or j in st:
                subtrees[subtrees.index(st)].append(i)
                subtrees[subtrees.index(st)].append(j)
                nodes.add(i), nodes.add(j)
    else:
        subtrees.append([i,j])
        nodes.add(i), nodes.add(j)

l = len(subtrees)
for i in range(l):
    for j in range(l):
        if i==j:
           break
        if len(set(subtrees[i] + subtrees[j])) < len(subtrees[i]) + len(subtrees[j]):
            subtrees[i] = list(set(subtrees[i] + subtrees[j]))
            subtrees[j] = []
subtrees = [i for i in subtrees if i]
result = (n -len(nodes)) + len(subtrees) - 1
print(result)
```

```python
#TRAN

from Bio import SeqIO

transition_list = [('A', 'G'), ('G', 'A'), ('T', 'C'), ('C', 'T')]
transversions_list = [('A', 'T'), ('T', 'A'), ('A', 'C'), ('C', 'A'), ('T', 'G'), ('G', 'T'), ('C', 'G'), ('G', 'C')]

def fastaReader(fasta_file):
    seq = []
    for x in SeqIO.parse(fasta_file, "fasta"):
        seq.append(str(x.seq))
    return seq

transition, transversions = 0, 0
seq = fastaReader("rosalind_tran.txt")

s1 = seq[0]
s2 = seq[1]

for i in range(len(s1)):
    if (s1[i], s2[i]) in transition_list:
        transition += 1
    if (s1[i], s2[i]) in transversions_list:
        transversions += 1
print(transition/transversions)
```
